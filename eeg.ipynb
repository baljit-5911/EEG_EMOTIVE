```python
from tensorflow.keras.layers import Conv1D, BatchNormalization, LeakyReLU, MaxPool1D, \
    GlobalAveragePooling1D, Dense, Dropout, AveragePooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.backend import clear_session

def cnnmodel():
    clear_session()
    model = Sequential()

    # First Conv Layer
    model.add(Conv1D(filters=8, kernel_size=3, strides=1, input_shape=(640, 14)))
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2, strides=2))

    # Second Conv Layer
    model.add(Conv1D(filters=8, kernel_size=3, strides=1))
    model.add(BatchNormalization())  # Added BatchNormalization to maintain consistency
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2, strides=2))
    model.add(Dropout(0.3))

    # Third Conv Layer
    model.add(Conv1D(filters=8, kernel_size=3, strides=1))
    model.add(BatchNormalization())  # Added BatchNormalization
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2, strides=2))
    model.add(Dropout(0.3))

    # Fourth Conv Layer
    model.add(Conv1D(filters=8, kernel_size=3, strides=1))
    model.add(BatchNormalization())  # Added BatchNormalization
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2, strides=2))

    # Fifth Conv Layer
    model.add(Conv1D(filters=8, kernel_size=3, strides=1))
    model.add(BatchNormalization())  # Added BatchNormalization
    model.add(LeakyReLU())

    # Global Average Pooling and Final Dense Layer for 4 Classes
    model.add(GlobalAveragePooling1D())
    model.add(Dense(5, activation='softmax'))  # Changed to 5 output classes

    # Compile the model for multi-class classification
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

model = cnnmodel()
model.summary()
```

```python
def classify_single_epoch(fif_file_path, model, scaler, epoch_idx=0):
    """
    Preprocess a single epoch of data from a .fif file and classify using the trained CNN model.

    Parameters:
        fif_file_path (str): Path to the .fif file.
        model (tf.keras.Model): Trained CNN model for classification.
        scaler (StandardScaler): Scaler fitted on training data to ensure consistent preprocessing.
        epoch_idx (int): Index of the epoch to process and classify.

    Returns:
        int: Predicted class for the selected epoch.
    """
    # Define EEG channels used during training
    ch_names = ['AF3', 'F7', 'F3', 'FC5', 'T7', 'P7', 'O1', 'O2', 'P8', 'T8', 'FC6', 'F4', 'F8', 'AF4']

    # Load and preprocess the .fif file
    try:
        raw_data = mne.io.read_raw_fif(fif_file_path, preload=True)
        raw_data.pick_channels(ch_names)  # Keep only required channels
    except Exception as e:
        raise ValueError(f"Error loading file {fif_file_path}: {str(e)}")

    # Segment the data into 5-second epochs
    epochs = mne.make_fixed_length_epochs(raw_data, duration=5, overlap=4).get_data()

    # Ensure the epoch index is valid
    if epoch_idx < 0 or epoch_idx >= epochs.shape[0]:
        raise IndexError(f"Epoch index {epoch_idx} is out of bounds. Total epochs: {epochs.shape[0]}")

    # Select the specified epoch
    single_epoch = epochs[epoch_idx]

    # Preprocess the epoch (scaling with the provided scaler)
    single_epoch = scaler.transform(single_epoch.T).T  # Transpose for scaling and then back

    # Adjust dimensions to match model input (samples, time, channels)
    single_epoch = np.expand_dims(single_epoch.T, axis=0)  # Shape: (1, time, channels)

    # Print the shape of the input
    print("Shape of input to the model:", single_epoch.shape)

    # Perform classification on the selected epoch
    prediction = model.predict(single_epoch)
    predicted_class = np.argmax(prediction, axis=1)[0]

    return predicted_class
# Assuming the scaler used during training is already available
predicted_class = classify_single_epoch(
    "/content/Real Time Task Classification - Copy/T2a/fif/20240909-015509-Epoc X-raw.fif",
    model,
    scaler,  # Use the training scaler here
    epoch_idx=153
)

print(f"Predicted class for epoch 0: {predicted_class}")
```
